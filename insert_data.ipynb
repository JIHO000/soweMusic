```python
song_list = []
for i in range(1,101):
    query.exec_("select songId from favorits where userId = '%s'"%(i))
    user_list = []
    while query.next():
        user_list.append(int(query.value(0)))
    song_list.append(user_list)
```

```python
import mlxtend
import numpy as np
import pandas as pd
```

```python
data = np.array(song_list)
df_data = pd.DataFrame(data)
df_data
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_array = te.fit(data).transform(data)

df = pd.DataFrame(te_array, columns=te.columns_)

df
```

```python
from mlxtend.frequent_patterns import apriori
apr = apriori(df, min_support=0.05,use_colnames=True)
apr
```

```python
apriori_list = []
for i in range(len(apr)):
    apriori_list.append(list(apr["itemsets"][i]))

user_items = [1, 5]

index = []
for i in range(len(apriori_list) - 1, -1, -1):
    if all(x in apriori_list[i] for x in user_items):
        print(apriori_list[i])
        recommend_list = apriori_list[i]
        for j in user_items:
            recommend_list.remove(j)
        break
    
```
